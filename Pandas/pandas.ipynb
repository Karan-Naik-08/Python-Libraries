## Parse Date Function

```python
# Sometimes Date gets converted to string so to avoidit use this function
df= pd.read_csv('nyc_weather.csv',parse_dates=['EST'])
```

# Handling Missing Values

## Fillna Function

```python
# Use fill na to fill all missing values 
df.fillna(0)

# To Fill Each Column With Different Values 
df.fillna({
    'EST' : 0,
    'Temperature' : 0,
    'Events' : 'Rain'
},inplace=True)
df.head(2)
```

```python
# Using Forward Fill, Here previous rows value is inserted in the next row
df.fillna(method='ffill')

# Backward fill, same as ff it , but uses next rows value
df.fillna(method='bfill')

# In both the cases we can use limit to set limit of how much rows to change ,like if limit is 1 only next 1 is filled with previous value rest vill remain null
df.shape
```

## Interpolate Function

```python
# This function uses liner method ( avg previous and next value to fill the null value).
df.interpolate()
df.head(10)

# This funcions default method is liner method but we can change it to Different methods like cubic, time etc.
```

## Drop Na

```python
# This function drops all the entire row if there's any missing value, but also we can specify on which column basis shall it drop using subset
df.dropna()
df.dropna(how='all') # Using this it will drop a row only when all values for cols are null

# And If we want to keep rows with atleast 1 non nanvalues use thresh 
df.dropna(thresh=1)
df.head(1)
```
